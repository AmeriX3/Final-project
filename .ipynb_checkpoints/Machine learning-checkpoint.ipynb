```python
cars= pd.read_csv("cars.csv")
cars.drop(columns="Unnamed: 0",inplace=True)
cars
```

```python
cor= cars.corr()
plt.figure(figsize=(12,10))
matrix = np.triu(cor)
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds, mask=matrix)
plt.show()
```

```python
def car4(s):
    if "5+" in s:
        return "4 door"
    else:
        pass
```

```python
carsd = pd.get_dummies(cars)
carsd
```

```python
x=carsd.drop(columns="car_price")
y=carsd["car_price"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
model = DecisionTreeRegressor()
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(carsd.drop(columns="car_price"))

lis = []
for i in predictions:
    lis.append(round(i,0))
cars["predictions"]=lis
cars
```

```python


r2_score(carsd["car_price"],predictions)
```
